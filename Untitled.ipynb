```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline



df = pd.read_csv("1632300362534233.csv")

# take a look at the dataset
df.head()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs[0, 0].scatter(df.Area, df.Price,  color='blue')
axs[0, 0].set_title('Area')

axs[0, 1].scatter(df.Room, df.Price,  color='blue')
axs[0, 1].set_title('Room')

axs[0, 2].scatter(df.Parking, df.Price,  color='blue')
axs[0, 2].set_title('Parking')

axs[1, 0].scatter(df.Warehouse, df.Price,  color='blue')
axs[1, 0].set_title('Warehouse')

axs[1, 1].scatter(df.Elevator, df.Price,  color='blue')
axs[1, 1].set_title('Elevator')


plt.show()
```

```python
# حذف خانه‌هایی که آدرس ندارند
df_cleaned = df.dropna(subset=['Address'])

# تبدیل متراژ به عددی به صورت امن با استفاده از loc
df_cleaned.loc[:, 'Area'] = pd.to_numeric(df_cleaned['Area'], errors='coerce')

# حذف خانه‌هایی که متراژ غیرمنطقی دارند (مثلاً زیر ۲۰ یا بالای ۱۰۰۰ متر مربع)
df_cleaned = df_cleaned[(df_cleaned['Area'] >= 20) & (df_cleaned['Area'] <= 1000)]


```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs[0, 0].scatter(df_cleaned.Area, df_cleaned.Price,  color='blue')
axs[0, 0].set_title('Area')

axs[0, 1].scatter(df_cleaned.Room, df_cleaned.Price,  color='blue')
axs[0, 1].set_title('Room')

axs[0, 2].scatter(df_cleaned.Parking, df_cleaned.Price,  color='blue')
axs[0, 2].set_title('Parking')

axs[1, 0].scatter(df_cleaned.Warehouse, df_cleaned.Price,  color='blue')
axs[1, 0].set_title('Warehouse')

axs[1, 1].scatter(df_cleaned.Elevator, df_cleaned.Price,  color='blue')
axs[1, 1].set_title('Elevator')

axs[1, 2].scatter(df_cleaned.Address, df_cleaned.Price,  color='blue')
axs[1, 2].set_title('Address')


plt.show()
```

```python
msk = np.random.rand(len(df_cleaned)) < 0.8
train = df_cleaned[msk]
test = df_cleaned[~msk]

# انتخاب x_data و y_data از train
x_data, y_data = (train["Room"].values, train["Price"].values)

# نرمال‌سازی داده‌ها
xdata = x_data / max(x_data)
ydata = y_data / max(y_data)

# تعریف تابع quadratic
def quadratic(x, a, b, c):
    y = a * np.power(x, 2) + b * x + c
    return y

# برازش منحنی
from scipy.optimize import curve_fit
popt, pcov = curve_fit(quadratic, xdata, ydata)

# چاپ پارامترهای نهایی
print(" a = %f, b = %f, c = %f" % (popt[0], popt[1], popt[2]))
```

```python
# داده‌های تست
x_test = test["Room"].values
y_test = test["Price"].values

# نرمال‌سازی داده‌های تست
x_test_norm = x_test / max(x_data)  # از max(x_data) که روی داده‌های آموزشی محاسبه شده بود، استفاده می‌کنیم

# پیش‌بینی با استفاده از مدل برازش‌شده
y_pred_norm = quadratic(x_test_norm, *popt)

# تبدیل پیش‌بینی‌ها به مقیاس اصلی
y_pred = y_pred_norm * max(y_data)  # مقادیر نرمال‌شده را به مقیاس اصلی برمی‌گردانیم

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# محاسبه خطاها
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5  # جذر برای محاسبه RMSE
r2 = r2_score(y_test, y_pred)

# چاپ نتایج
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Square Error (RMSE): {rmse}")
print(f"R-squared (R²): {r2}")
```

```python

# رسم نمودار
plt.figure(figsize=(8, 5))

# نرمال‌سازی داده‌های آموزشی و تست
x_train_norm = train["Room"].values / max(x_data)
y_train_norm = train["Price"].values / max(y_data)
x_test_norm = test["Room"].values / max(x_data)
y_test_norm = test["Price"].values / max(y_data)

# رسم داده‌های آموزشی (Train) با نقاط قرمز
plt.scatter(x_train_norm, y_train_norm, color='red', label='Train Data')

# رسم داده‌های تست (Test) با نقاط آبی
plt.scatter(x_test_norm, y_test_norm, color='blue', label='Test Data')

# رسم خط برازش‌شده براساس ضرایب a، b، c
x_line = np.linspace(0, 1, 100)  # تولید مقادیر x برای رسم خط برازش‌شده
y_line = quadratic(x_line, *popt)  # محاسبه مقادیر y براساس مدل برازش‌شده

# رسم خط برازش‌شده
plt.plot(x_line, y_line, color='green', linewidth=3, label='Fitted Curve')

# اضافه کردن عنوان و عناوین محورهای نمودار
plt.xlabel('Room (Normalized)')
plt.ylabel('Price (Normalized)')
plt.legend(loc='best')

# نمایش نمودار
plt.show()
```
