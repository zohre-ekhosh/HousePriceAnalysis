```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Ø¨Ø§Ø±Ú¯Ø°Ø§Ø±Û Ø¯Ø§Ø¯ÙâÙØ§
df = pd.read_csv("1632300362534233.csv")

# Ø­Ø°Ù Ø®Ø§ÙÙâÙØ§ÛÛ Ú©Ù Ø¢Ø¯Ø±Ø³ ÙØ¯Ø§Ø±ÙØ¯
df_cleaned = df.dropna(subset=['Address'])

# ØªØ¨Ø¯ÛÙ ÙØªØ±Ø§Ú Ø¨Ù Ø¹Ø¯Ø¯Û
df_cleaned.loc[:, 'Area'] = pd.to_numeric(df_cleaned['Area'], errors='coerce')

# Ø­Ø°Ù Ø®Ø§ÙÙâÙØ§ÛÛ Ú©Ù ÙØªØ±Ø§Ú ØºÛØ±ÙÙØ·ÙÛ Ø¯Ø§Ø±ÙØ¯
df_cleaned = df_cleaned[(df_cleaned['Area'] >= 20) & (df_cleaned['Area'] <= 1000)]

# ØªÙØ³ÛÙ Ø¯Ø§Ø¯ÙâÙØ§ Ø¨Ù Ø¯Ø§Ø¯ÙâÙØ§Û Ø¢ÙÙØ²Ø´Û Ù ØªØ³Øª
msk = np.random.rand(len(df_cleaned)) < 0.8
train = df_cleaned[msk]
test = df_cleaned[~msk]

# Ø§ÙØªØ®Ø§Ø¨ x_data Ù y_data Ø§Ø² train
x1_data = train["Room"].values
x2_data = train["Area"].values
y_data = train["Price"].values

# ÙØ±ÙØ§ÙâØ³Ø§Ø²Û Ø¯Ø§Ø¯ÙâÙØ§
x1data = x1_data / max(x1_data)
x2data = x2_data / max(x2_data)
ydata = y_data / max(y_data)

# ØªØ¹Ø±ÛÙ ØªØ§Ø¨Ø¹ Ø±Ú¯Ø±Ø³ÛÙÙ ÚÙØ¯ ÙØªØºÛØ±Ù
def quadratic_multivariate(x, a, b, c, d):
    x1, x2 = x  # Ø§Ø³ØªØ®Ø±Ø§Ø¬ x1 Ù x2 Ø§Ø² ÙØ±ÙØ¯Û
    return a * np.power(x1, 2) + b * x1 + c * x2 + d

# Ø¨Ø±Ø§Ø²Ø´ ÙÙØ­ÙÛ
popt, pcov = curve_fit(quadratic_multivariate, 
                       (x1data, x2data), ydata)

# ÚØ§Ù¾ Ù¾Ø§Ø±Ø§ÙØªØ±ÙØ§Û ÙÙØ§ÛÛ
print(" a = %f, b = %f, c = %f, d = %f" % (popt[0], popt[1], popt[2], popt[3]))

# Ø¯Ø§Ø¯ÙâÙØ§Û ØªØ³Øª
x1_test = test["Room"].values
x2_test = test["Area"].values
y_test = test["Price"].values

# ÙØ±ÙØ§ÙâØ³Ø§Ø²Û Ø¯Ø§Ø¯ÙâÙØ§Û ØªØ³Øª
x1_test_norm = x1_test / max(x1_data)
x2_test_norm = x2_test / max(x2_data)

# Ù¾ÛØ´âØ¨ÛÙÛ Ø¨Ø§ Ø§Ø³ØªÙØ§Ø¯Ù Ø§Ø² ÙØ¯Ù Ø¨Ø±Ø§Ø²Ø´âØ´Ø¯Ù
y_pred_norm = quadratic_multivariate((x1_test_norm, x2_test_norm), *popt)

# ØªØ¨Ø¯ÛÙ Ù¾ÛØ´âØ¨ÛÙÛâÙØ§ Ø¨Ù ÙÙÛØ§Ø³ Ø§ØµÙÛ
y_pred = y_pred_norm * max(y_data)

# ÙØ­Ø§Ø³Ø¨Ù Ø®Ø·Ø§ÙØ§
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5
r2 = r2_score(y_test, y_pred)

# ÚØ§Ù¾ ÙØªØ§ÛØ¬
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Square Error (RMSE): {rmse}")
print(f"R-squared (RÂ²): {r2}")

# Ø±Ø³Ù ÙÙÙØ¯Ø§Ø±
plt.figure(figsize=(8, 5))

# ÙØ±ÙØ§ÙâØ³Ø§Ø²Û Ø¯Ø§Ø¯ÙâÙØ§Û Ø¢ÙÙØ²Ø´Û Ù ØªØ³Øª
x1_train_norm = train["Room"].values / max(x1_data)
x2_train_norm = train["Area"].values / max(x2_data)

# Ø±Ø³Ù Ø¯Ø§Ø¯ÙâÙØ§Û Ø¢ÙÙØ²Ø´Û (Train) Ø¨Ø§ ÙÙØ§Ø· ÙØ±ÙØ²
plt.scatter(x1_train_norm, ydata, color='red', label='Train Data')

# Ø±Ø³Ù Ø¯Ø§Ø¯ÙâÙØ§Û ØªØ³Øª (Test) Ø¨Ø§ ÙÙØ§Ø· Ø¢Ø¨Û
plt.scatter(x1_test_norm, y_test / max(y_data), color='blue', label='Test Data')

# Ø±Ø³Ù Ø®Ø· Ø¨Ø±Ø§Ø²Ø´âØ´Ø¯Ù
x1_line = np.linspace(0, 1, 100)  # ØªÙÙÛØ¯ ÙÙØ§Ø¯ÛØ± x Ø¨Ø±Ø§Û Ø±Ø³Ù Ø®Ø· Ø¨Ø±Ø§Ø²Ø´âØ´Ø¯Ù
x2_line = np.mean(x2data)  # ÙÛâØªÙØ§ÙÛØ¯ ÛÚ© ÙÙØ¯Ø§Ø± Ø«Ø§Ø¨Øª ÛØ§ ÙÛØ§ÙÚ¯ÛÙ Ø§Ø³ØªÙØ§Ø¯Ù Ú©ÙÛØ¯
y_line = quadratic_multivariate((x1_line, x2_line), *popt)  # ÙØ­Ø§Ø³Ø¨Ù ÙÙØ§Ø¯ÛØ± y Ø¨Ø±Ø§Ø³Ø§Ø³ ÙØ¯Ù Ø¨Ø±Ø§Ø²Ø´âØ´Ø¯Ù

# Ø±Ø³Ù Ø®Ø· Ø¨Ø±Ø§Ø²Ø´âØ´Ø¯Ù
plt.plot(x1_line, y_line, color='green', linewidth=3, label='Fitted Curve')

# Ø§Ø¶Ø§ÙÙ Ú©Ø±Ø¯Ù Ø¹ÙÙØ§Ù Ù Ø¹ÙØ§ÙÛÙ ÙØ­ÙØ±ÙØ§Û ÙÙÙØ¯Ø§Ø±
plt.xlabel('Room (Normalized)')
plt.ylabel('Price (Normalized)')
plt.legend(loc='best')

# ÙÙØ§ÛØ´ ÙÙÙØ¯Ø§Ø±
plt.show()
```
