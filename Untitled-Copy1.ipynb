```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# بارگذاری داده‌ها
df = pd.read_csv("1632300362534233.csv")

# حذف خانه‌هایی که آدرس ندارند
df_cleaned = df.dropna(subset=['Address'])

# تبدیل متراژ به عددی
df_cleaned.loc[:, 'Area'] = pd.to_numeric(df_cleaned['Area'], errors='coerce')

# حذف خانه‌هایی که متراژ غیرمنطقی دارند
df_cleaned = df_cleaned[(df_cleaned['Area'] >= 20) & (df_cleaned['Area'] <= 1000)]

# تقسیم داده‌ها به داده‌های آموزشی و تست
msk = np.random.rand(len(df_cleaned)) < 0.8
train = df_cleaned[msk]
test = df_cleaned[~msk]

# انتخاب x_data و y_data از train
x1_data = train["Room"].values
x2_data = train["Area"].values
y_data = train["Price"].values

# نرمال‌سازی داده‌ها
x1data = x1_data / max(x1_data)
x2data = x2_data / max(x2_data)
ydata = y_data / max(y_data)

# تعریف تابع رگرسیون چند متغیره
def quadratic_multivariate(x, a, b, c, d):
    x1, x2 = x  # استخراج x1 و x2 از ورودی
    return a * np.power(x1, 2) + b * x1 + c * x2 + d

# برازش منحنی
popt, pcov = curve_fit(quadratic_multivariate, 
                       (x1data, x2data), ydata)

# چاپ پارامترهای نهایی
print(" a = %f, b = %f, c = %f, d = %f" % (popt[0], popt[1], popt[2], popt[3]))

# داده‌های تست
x1_test = test["Room"].values
x2_test = test["Area"].values
y_test = test["Price"].values

# نرمال‌سازی داده‌های تست
x1_test_norm = x1_test / max(x1_data)
x2_test_norm = x2_test / max(x2_data)

# پیش‌بینی با استفاده از مدل برازش‌شده
y_pred_norm = quadratic_multivariate((x1_test_norm, x2_test_norm), *popt)

# تبدیل پیش‌بینی‌ها به مقیاس اصلی
y_pred = y_pred_norm * max(y_data)

# محاسبه خطاها
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5
r2 = r2_score(y_test, y_pred)

# چاپ نتایج
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Square Error (RMSE): {rmse}")
print(f"R-squared (R²): {r2}")

# رسم نمودار
plt.figure(figsize=(8, 5))

# نرمال‌سازی داده‌های آموزشی و تست
x1_train_norm = train["Room"].values / max(x1_data)
x2_train_norm = train["Area"].values / max(x2_data)

# رسم داده‌های آموزشی (Train) با نقاط قرمز
plt.scatter(x1_train_norm, ydata, color='red', label='Train Data')

# رسم داده‌های تست (Test) با نقاط آبی
plt.scatter(x1_test_norm, y_test / max(y_data), color='blue', label='Test Data')

# رسم خط برازش‌شده
x1_line = np.linspace(0, 1, 100)  # تولید مقادیر x برای رسم خط برازش‌شده
x2_line = np.mean(x2data)  # می‌توانید یک مقدار ثابت یا میانگین استفاده کنید
y_line = quadratic_multivariate((x1_line, x2_line), *popt)  # محاسبه مقادیر y براساس مدل برازش‌شده

# رسم خط برازش‌شده
plt.plot(x1_line, y_line, color='green', linewidth=3, label='Fitted Curve')

# اضافه کردن عنوان و عناوین محورهای نمودار
plt.xlabel('Room (Normalized)')
plt.ylabel('Price (Normalized)')
plt.legend(loc='best')

# نمایش نمودار
plt.show()
```
